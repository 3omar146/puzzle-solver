Provide image path in the next cell.

```python
import cv2, matplotlib.pyplot as plt

IMAGE_PATH = "data/raw/Gravity Falls/puzzle_2x2/1.jpg"   #  CHANGE THIS
original = cv2.imread(IMAGE_PATH)
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.title('Original');
```

```python
from src.enhancement import to_grayscale, denoise_bilateral, apply_clahe, sharpen
import matplotlib.pyplot as plt

# Grayscale
gray = to_grayscale(original)
plt.imshow(gray, cmap='gray'); plt.title('Grayscale'); plt.axis('off')
```

```python
# Bilateral Denoising
den = denoise_bilateral(gray)
plt.imshow(den, cmap='gray'); plt.title('Bilateral'); plt.axis('off')
```

```python
# CLAHE
cla = apply_clahe(den)
plt.imshow(cla, cmap='gray'); plt.title('CLAHE'); plt.axis('off')
```

```python
# Sharpen (No CLAHE)
sharp = sharpen(den)
plt.imshow(sharp, cmap='gray'); plt.title('Sharpened'); plt.axis('off')
```

```python
# Sharpen (CLAHE)
sharp_c = sharpen(cla)
plt.imshow(sharp_c, cmap='gray'); plt.title('Sharpened CLAHE'); plt.axis('off')
```
